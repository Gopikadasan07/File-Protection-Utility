```python
import base64

print("1. Protect Data")
print("2. Restore Data")

choice = input("Enter your choice: ")

if choice == "1":
    text = input("Enter text to protect: ")
    protected = base64.b64encode(text.encode()).decode()
    print("Protected Data:", protected)

elif choice == "2":
    protected = input("Enter protected data: ")
    restored = base64.b64decode(protected.encode()).decode()
    print("Restored Data:", restored)

else:
    print("Invalid choice")
```
